```python
# import the library
import folium
import pandas as pd
from geopy.geocoders import Nominatim
import os
import glob

## get container locations
df = pd.read_csv ('../data/raw/Container_Locations.csv', sep=";")
del df['Sens. Typ']
del df['Cont. Typ']
del df['Farbe']
df.rename(columns = {'Sensor':'Sensor'})
geolocator = Nominatim(user_agent="http")

# initialize map
m = folium.Map(location=[50.110924,8.682127], tiles="OpenStreetMap", zoom_start=12)
```

```python
# Cluster 0
path0 = "../data/preprocessed/clusters/0"
csv_files0 = glob.glob(os.path.join(path0, "*.csv"))

sensorList0 = []

for f in csv_files0: 
    dfCluster0 = pd.read_csv(f)
    sensorId0 = (f.split("\\")[-1]).replace(".csv","")
    sensorList0.append(sensorId0)
    
    
print(sensorList0)
```

```python
dfs0 = []

for index, row in df.iterrows():
    Sensor0 = row['Sensor']
    if Sensor0 in sensorList0:
        try:
            location0 = geolocator.geocode(row['Location'] + ', Frankfurt')
            new_row0 = {'lon': location0.longitude, 'lat': location0.latitude, 'name': Sensor0,}
            dfs0.append(new_row0)
            print (new_row0)
        except:
            print ('Error, nvm')
        

data0 = pd.DataFrame(dfs0, columns =['lon', 'lat', 'name' ])
print(data0)

for i in range(0,len(data0)):
   folium.Marker(
      location=[data0.iloc[i]['lat'], data0.iloc[i]['lon']],
      popup=data0.iloc[i]['name'],
      icon=folium.Icon(color="red",  icon="fa-trash-o", prefix="fa")
       
   ).add_to(m)
```
