```python
#get data from cluster 0

columns = ['id','deveui', 'unix_time', 'client_id', 'created_at', 'Status', 'Sensor ID', 'Events', 'Height', 'Voltage', 'Temperature', 'Tilt', 'Tx Event', 'Messagetype']
temp = pd.DataFrame(columns=columns)
path_to_csv = '../data/preprocessed/clusters/0/' 

#use example fill process

csv_files = [pos_csv for pos_csv in os.listdir(path_to_csv) if pos_csv.endswith('F.csv') if not pos_csv.startswith('7') ]



dfs0 = []

for filename in csv_files:
    df = pd.read_csv(path_to_csv + filename, index_col=0)
    df
    dfs0.append(df)

     
temp0 = pd.concat(dfs0)
del temp0['Tilt']
del temp0['lfilter']
del temp0['Temperature']
del temp0['mov_avg']
del temp0['min_avg']
del temp0['Height']
temp0.index = pd.to_datetime(temp0.index)
temp0.index= temp0.index.map(dt.datetime.toordinal)
# temp
cluster0 = temp0.head(13)

```
